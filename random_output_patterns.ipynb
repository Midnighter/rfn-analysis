```python
import numpy
import sys
```

```python
class Parameters(object):
    def __init__(self):
        self.n_in = 8
        self.m = 20
        self.n_out = 8
        self.k = 4
        self.h = 0.007
```

```python
def random_pattern(parameters):
    rand_int = numpy.random.randint
    pattern = numpy.zeros(shape=(parameters.n_out, parameters.n_in), dtype=float)
    for j in range(parameters.n_in):
        pos = set()
        while len(pos) < parameters.k:
            pos.add(rand_int(parameters.n_out))
        pattern[list(pos), j] = rand_int(sys.maxint / parameters.k, size=parameters.k)
    pattern /= pattern.sum(axis=0)
    return pattern
```

```python
def plot_pattern(matrix):
    plt.matshow(matrix, cmap=plt.cm.gray_r, vmin=0.0, vmax=1.0)
    plt.colorbar()
    plt.show()
```

```python
parameters = Parameters()
parameters.n_in = 4
```

```python
pattern = random_pattern(parameters)
plot_pattern(pattern)
```

```python
numpy.linalg.matrix_rank(pattern)
```

```python
for i in xrange(1000):
    pattern = random_pattern(parameters)
    binary_pattern = (pattern > 0.0)
    rank = numpy.linalg.matrix_rank(binary_pattern)
    assert rank == 4, "binary rank actually is: %d" % rank
```
